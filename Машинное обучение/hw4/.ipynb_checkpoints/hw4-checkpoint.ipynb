```python
import numpy as np
import pandas as pd
```

```python
from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline
```

Загрузим данные

```python
data = pd.read_excel('geo.xlsx')
```

```python
data.head()
```

```python
x = np.array(data.x)
y = np.array(data.y)
```

Нарисуем точки голосования на графике

```python
plt.figure(figsize = [8, 6])
plt.scatter(x.T, y.T)
```

Приведём данные по обеим осям к нормальному распределению

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X = np.array([x, y])
y = np.array(data.comment_class)
```

```python
X_scaled = scaler.fit_transform(X.T)
```

```python
plt.figure(figsize = [8, 6])
plt.scatter(X_scaled[:, 0], X_scaled[:, 1])
```

Импортируем библиотеки с различными методами кластеризации

```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, AffinityPropagation
from sklearn.metrics import adjusted_rand_score, silhouette_score
```

```python
model1 = DBSCAN()
```

```python
clusters1 = model1.fit_predict(X_scaled)
```

```python
plt.figure(figsize = [8, 6])
plt.scatter(data.x, data.y, c=clusters1)
```

```python
adjusted_rand_score(y, clusters1)
```

DBSCAN плохо справляется с кластеризацией, так как точки распложены плотно

```python
model2 = KMeans(n_clusters=7)
```

```python
clusters2 = model2.fit_predict(X_scaled)
```

```python
plt.figure(figsize = [8, 6])
plt.scatter(data.x, data.y, c=clusters2)
```

```python
adjusted_rand_score(y, clusters2)
```

K Means лучше справляется с определением районов внутри города, но не идеально

```python
model3 = AgglomerativeClustering(n_clusters=10)
```

```python
clusters3 = model3.fit_predict(X_scaled)
```

```python
plt.figure(figsize = [8, 6])
plt.scatter(data.x, data.y, c=clusters3)
```

```python
adjusted_rand_score(y, clusters3)
```

AgglomerativeCLustering не работает

```python
model4 = AffinityPropagation()
```

```python
cluster4 = model4.fit_predict(X_scaled)
```

```python
plt.figure(figsize = [8, 6])
plt.scatter(data.x, data.y, c=clusters4)
```

```python
adjusted_rand_score(y, clusters4)
```

Affinity propagtion выдает memory error

Теперь выведем точки на карте, где и как голосовали

```python
plt.figure(figsize = [8, 6])
plt.scatter(data.x, data.y, c=data.comment_class)
```

На карте видно, что тип голоса расположен на карте хаотично и нет возможности каким бы то ни было способом с помощью кластеризации обнаружить где и как голосовали граждане.
