```python
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
sns.set()
```

### Часть 1

Матрица переходных вероятностей

```python
with open('tom_1.txt', 'r', encoding='utf-8') as f:
    text = f.read().lower()
text = re.sub('[^а-яё]','', text)
```

```python
RUSSIAN = 'абвгдеёжзийклмнопрстуфхцчшщъыьэюя'
POS = {l: i for i, l in enumerate(RUSSIAN)}
```

```python
probabilities = np.zeros((len(RUSSIAN), len(RUSSIAN)))
```

```python
for cp, cn in zip(text[:-1], text[1:]):
    probabilities[POS[cp], POS[cn]] += 1
```

```python
probabilities = (probabilities / probabilities.sum(axis=1)[:, None])
```

```python
df = pd.DataFrame(probabilities, index=(c for c in RUSSIAN), columns=(c for c in RUSSIAN))
```

```python
plt.figure(figsize=(14,8))
sns.heatmap(df)
plt.show()
```

Вероятность встретить строку из 4-х символов, которая начинается п, а кончается на р

```python
start = POS['п']
end = POS['р']
prior = 1. / len(RUSSIAN)
total_proba = 0
for c1 in RUSSIAN:
    for c2 in RUSSIAN:
        proba = (
            prior 
            * probabilities[start, POS[c1]] 
            * probabilities[POS[c1], POS[c2]]
            * probabilities[POS[c2], end]
        )
        if proba > 0:
            total_proba += proba
print('Вероятность:', total_proba)
```

Используя матрицу переходных вероятностей сгенерировать слово, начинающееся с буквы “Б” и имеющее длину 8-мь символов

```python
s = []
start = 'б'
s.append(start)
length = 8
for i in range(length):
    index = POS[s[i]]
    next_char = np.random.choice(list(RUSSIAN), p=probabilities[index])
    s.append(next_char)

print (''.join(s))
```

### Часть 2

```python
from hmmlearn.hmm import GaussianHMM
```

```python
series = pd.read_csv("weekly-closings-of-the-dowjones-.csv")["Close"]
```

```python
values = series.values.reshape(-1,1)
```

```python
model = GaussianHMM(n_components=9,
                        covariance_type="diag",
                        n_iter=1000)
model.fit(values)
labels = model.predict(values)
means = np.zeros_like(values)
for i in range(model.n_components):
    means[labels == i] = model.means_[i]
    
plt.figure(figsize=(12, 6))
plt.plot(values)
plt.plot(means, linewidth=3)
plt.show()
```

```python
def predict(series, N, n_pred):
    new_series = series.copy()
    for _ in range(n_pred):
        new_series = new_series.append(pd.Series([new_series[-N:].mean()]), ignore_index=True)
    return new_series
```

```python
new_series = predict(series, 7, 20)
```

```python
new_series = series.append(new_series)
```

```python
new_values = new_series.values.reshape(-1,1)
model = GaussianHMM(n_components=9,
                        covariance_type="diag",
                        n_iter=1000)
model.fit(new_values)
labels = model.predict(new_values)
means = np.zeros_like(new_values)
for i in range(model.n_components):
    means[labels == i] = model.means_[i]
    
plt.figure(figsize=(12, 6))
plt.plot(new_values)
plt.plot(means, linewidth=3)
plt.show()
```
