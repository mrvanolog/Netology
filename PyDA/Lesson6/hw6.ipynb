```python
import pandas as pd
```

```python
rating = pd.read_csv('ratings.csv')
movie = pd.read_csv('movies.csv')
movie.head()
```

```python
five_star = rating.query('rating == 5.0')
top_rating = five_star.movieId.value_counts(ascending=False).head(1)

top_movie = movie.query(f'movieId == {top_rating.index[0]}')

top_movie
```

**Задание 2**

```python
data = pd.read_csv('power.csv')

data.head()
```

```python
baltic = data[ (data['country'] == 'Lithuania') 
                | (data['country'] == 'Latvia') 
                | (data['country'] == 'Estonia') ]

year = baltic[ (baltic['year'] >= 2005) & (baltic['year'] <= 2010) ]
              
category = year[ (year['category'] == 4) 
                  | (year['category'] == 12)
                  | (year['category'] == 21)]

#category.head(100)

new_data = category[ category['quantity'] >= 0 ]

quantity_sum = new_data.sum(axis = 0, skipna = True)[2]

print("Суммарное потребление стран Прибалтики:", quantity_sum)
```
