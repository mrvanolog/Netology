**Задание 1**

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
ratings_grouped = ratings.groupby('userId').count()
ratings_filtered = ratings_grouped[ratings_grouped['rating'] > 99].reset_index()
fans_list = ratings_filtered['userId'].tolist()
ratings = ratings[ ratings['userId'].isin(fans_list) ]
ratings.head(10)
```

```python
subtracted = ratings.groupby('userId').max() - ratings.groupby('userId').min()
lifetime = subtracted['timestamp'].mean()
print(f'Mean lifetime is: {lifetime} ms or {lifetime / 3600 / 24} days')
```
