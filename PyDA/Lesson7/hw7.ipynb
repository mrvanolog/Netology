**Задание 1**

```python
def rate(rating):
    if rating <= 2:
        return 'низкий рейтинг'
    elif rating <= 4:
        return 'средний рейтинг'
    else:
        return 'высокий рейтинг'
```

```python
rating = pd.read_csv('ml-latest-small/ratings.csv')
rating['class'] = rating.rating.apply(rate)
rating.head(10)
```

**Задание 2**

```python
geo_data = {'Центр': ['москва', 'тула', 'ярославль'],
            'Северо-Запад': ['петербург', 'псков', 'мурманск'],
            'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']}

keywords = pd.read_csv('keywords.csv')
keywords.head()
```

```python
def geo_class(keywords, geo_data):
    words = keywords.split(' ')
    
    for word in words:
        for key in geo_data:
            if word in geo_data[key]:
                return key
    
    return 'undefined'
```

```python
keywords['region'] = keywords.keyword.apply(geo_class, geo_data = geo_data)
keywords.head(10)
```

**Задание 3**

```python
years = np.arange(1950, 2011)
ratings = pd.read_csv('ml-latest-small/ratings.csv')
movies = pd.read_csv('ml-latest-small/movies.csv')
```

```python
def production_year(string, years):
    for year in years:
        if f'{year}' in string:
            return year
    return 1900
```

```python
data = pd.merge(movies, ratings, on='movieId')
data.drop(['movieId', 'genres', 'userId', 'timestamp'], axis=1, inplace=True)
data['year'] = data['title'].apply(production_year, years=years)
```

```python
gd = data.groupby('year').mean()

gd.sort_values('rating', ascending=False).head(10)
```
